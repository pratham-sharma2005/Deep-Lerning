```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
img = image.load_img("/Users/prathamsharma/Desktop/cat/cat1.jpeg",target_size=(200,200))
```

```python
plt.imshow(img)
```

```python
img
```

```python
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest' #fill the empty pixels generated after shifting by the nearest pixels other options are "constant" ,"reflect"
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i = 0 

for output in datagen.flow( input_batch , batch_size=1 , save_to_dir="aug"):
    i = i + 1
    if i == 10:
        break
```
